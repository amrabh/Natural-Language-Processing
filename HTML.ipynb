```python
import nltk
from urllib.request import urlopen 
```

```python
url = "https://en.wikipedia.org/wiki/Python_(programming_language)"
```

```python
html = urlopen(url).read()
```

```python
html
```

```python
from bs4 import BeautifulSoup
```

```python
web_str = BeautifulSoup(html).get_text()
```

```python
web_str
```

```python
start = web_str.find("Python is an interpreted high-level general-purpose programming language.")
```

```python
start
```

```python
end = web_str.find("Python consistently ranks as one of the most popular programming languages")
```

```python
end
```

```python
last_sen = len("Python consistently ranks as one of the most popular programming languages")
```

```python
intro = web_str[start:end+last_sen]
```

```python
intro_tokens = nltk.word_tokenize(intro)
```

```python
print(intro_tokens)
```
