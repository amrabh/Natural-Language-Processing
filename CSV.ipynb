```python
import nltk
```

```python
import csv
```

```python
comments = []
with open("reviews.csv", "r") as file:
    reader = csv.reader(file)
    for row in reader:
        comments.append(row)
file.close()
```

```python
len(comments)
```

```python
comments[0]
```

```python
tokens = [nltk.word_tokenize(str(entry)) for entry in comments]
```
